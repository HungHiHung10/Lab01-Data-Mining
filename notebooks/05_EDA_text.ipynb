## 4. Tính và vẽ phân phối **độ dài văn bản (số từ, số ký tự)** theo nhãn lớp. Kiểm định Mann-Whitney U test để xác định liệu các lớp có sự khác biệt đáng kể về độ dài không.

Dưới góc độ Khoa học Dữ liệu (Data Science), việc phân tích độ dài văn bản và kiểm định thống kê là những bước đi mang tính nền tảng trong quá trình Khám phá Dữ liệu (EDA). Quá trình này giúp ta hiểu rõ những tính chất đặc điểm của ngôn ngữ để phân tích trước khi thực hiện làm đầu vào huấn luyện cho các mô hình phức tạp.

### Cơ sở Lý thuyết về Phân phối Độ dài Văn bản

Trong Xử lý Ngôn ngữ Tự nhiên (NLP), độ dài của một văn bản mang lại những manh mối quan trọng về hành vi của người dùng (ví dụ: người tức giận thường viết rất ngắn hoặc rất dài). Chúng ta thường đo lường độ dài qua 2 chiều kích thước:

#### 1. Số lượng từ (Word Count)
* **Cơ sở lý thuyết:** Đo lường số lượng "đơn vị có nghĩa" (tokens) trong một văn bản. Số từ phản ánh độ chi tiết của thông tin được truyền đạt.
* **Định nghĩa toán học:** Cho một văn bản (document) $d$ được cấu thành từ một tập hợp các từ (tokens) sau quá trình tách từ (tokenization): $d = \{w_1, w_2, \dots, w_n\}$.
Độ dài theo số từ của văn bản $d$ được tính bằng lực lượng của tập hợp đó:
$$W(d) = |d| = n$$

#### 2. Số lượng ký tự (Character Length)
* **Cơ sở lý thuyết:** Đo lường tổng số ký hiệu (bao gồm cả chữ cái, số, dấu câu và khoảng trắng). Mặc dù có độ tương quan mạnh với số từ, số ký tự giúp phát hiện các mẫu dữ liệu bất thường (Outliers) như việc lạm dụng dấu chấm than ("!!!!!"), lỗi gõ phím lặp lại ("so gooooood"), hoặc các đoạn mã nhiễu HTML.
* **Định nghĩa toán học:** Cho văn bản $d$ là một chuỗi (string) gồm $k$ ký tự $c$: $d = c_1 c_2 \dots c_k$.
Độ dài theo số ký tự của văn bản $d$ là:
$$C(d) = \sum_{i=1}^{k} 1 = k$$

#### 3. Công cụ Trực quan hóa (Visualization Theory)
Để quan sát "phân phối" (distribution) của các đại lượng này, ta sử dụng:
* **Histogram & KDE (Kernel Density Estimation):** Nhằm hỗ trợ để quan sát hình dáng tổng thể. Trong NLP, phân phối này thường **không tuân theo luật phân phối chuẩn (Non-normal)** mà mang hình dáng **lệch phải mạnh (Right-skewed / Power-law behavior)**. Nghĩa là phần lớn văn bản có độ dài ngắn, và chỉ có một cái "đuôi" dài rải rác các văn bản cực kỳ dài.
* **Boxplot (Biểu đồ hộp):** Công cụ đắc lực để quan sát sự tập trung ở vùng lõi (IQR - Interquartile Range) và so sánh trực diện trung vị (Median) giữa các lớp (Positive vs. Negative), đồng thời phát hiện các điểm ngoại lai (Outliers).


### Cơ sở Lý thuyết Kiểm định Mann-Whitney U Test

Khi ta nhìn thấy độ dài trung bình (Mean) của lớp Positive lớn hơn lớp Negative, ta không thể kết luận ngay là "có sự khác biệt". Sự chênh lệch đó có thể chỉ là do ngẫu nhiên của việc lấy mẫu. Do đó, ta cần một công cụ thống kê suy diễn để xác nhận.

#### 1. Tại sao lại chọn Mann-Whitney U Test?
* Trong thống kê cổ điển, để so sánh trung bình của 2 nhóm độc lập, người ta thường dùng **Independent T-test**.
* **Tuy nhiên**, T-test yêu cầu một giả định bắt buộc: **Dữ liệu phải có phân phối chuẩn (Normal Distribution)**. Như đã phân tích ở trên, độ dài văn bản luôn bị lệch phải nghiêm trọng (Right-skewed).
* **Giải pháp:** **Mann-Whitney U Test** (hay Wilcoxon rank-sum test) là một **kiểm định phi tham số (Non-parametric test)**. Nó không yêu cầu dữ liệu phải có phân phối chuẩn, vì nó không so sánh giá trị trung bình (Mean) mà so sánh **thứ hạng (Ranks)** của dữ liệu. Do đó, đây là "tiêu chuẩn vàng" (Gold Standard) cho bài toán này.

#### 2. Giả thuyết Thống kê (Statistical Hypotheses)
* **Giả thuyết không ($H_0$):** Không có sự khác biệt về phân phối độ dài văn bản giữa hai quần thể (Lớp Positive và Lớp Negative). Xác suất để một văn bản lấy ngẫu nhiên từ lớp này dài hơn lớp kia là 50%.
* **Giả thuyết thay thế ($H_1$):** Có sự khác biệt có ý nghĩa thống kê về phân phối độ dài văn bản giữa hai lớp.

#### 3. Các bước tính toán và Công thức Toán học
Giả sử ta có hai nhóm mẫu độc lập: Nhóm 1 (kích thước $n_1$) và Nhóm 2 (kích thước $n_2$).

* **Bước 1: Gộp và Xếp hạng.** Gộp toàn bộ $(n_1 + n_2)$ quan sát lại với nhau và xếp hạng từ bé đến lớn (giá trị nhỏ nhất có hạng 1, lớn nhất có hạng $n_1 + n_2$). Nếu có các giá trị bằng nhau (Ties), ta lấy trung bình cộng của các thứ hạng đó.
* **Bước 2: Tính tổng hạng.** Tính tổng các thứ hạng của riêng Nhóm 1 (ký hiệu là $R_1$) và Nhóm 2 (ký hiệu là $R_2$).
* **Bước 3: Tính thống kê U.** Đại lượng U cho từng nhóm được tính bằng công thức:
    $$U_1 = n_1 n_2 + \frac{n_1(n_1 + 1)}{2} - R_1$$
    $$U_2 = n_1 n_2 + \frac{n_2(n_2 + 1)}{2} - R_2$$
    *(Bản chất của phép toán này là đếm xem có bao nhiêu lần một giá trị của nhóm này lớn hơn một giá trị của nhóm kia).*
* **Bước 4: Xác định giá trị U cuối cùng.** Giá trị thống kê kiểm định $U$ là giá trị nhỏ hơn trong hai giá trị trên:
$$U = \min(U_1, U_2)$$
* **Bước 5: Tính P-value và Kết luận.**
    Đối với các mẫu có kích thước lớn ($n_1, n_2 > 20$), phân phối của $U$ sẽ xấp xỉ phân phối chuẩn. Thống kê $Z$ được tính như sau:
    $$Z = \frac{U - m_U}{\sigma_U}$$
    Trong đó, Kỳ vọng $m_U$ và Độ lệch chuẩn $\sigma_U$ của thống kê U là:
    $$m_U = \frac{n_1 n_2}{2}$$
    $$\sigma_U = \sqrt{\frac{n_1 n_2 (n_1 + n_2 + 1)}{12}}$$
    Từ giá trị $Z$, ta tra bảng phân phối chuẩn tắc để tìm ra **$p$-value**.

#### 4. Quy tắc đưa ra quyết định
Cho trước mức ý nghĩa thống kê $\alpha$ (thông thường chọn $\alpha = 0.05$):
* Nếu **$p$-value < $\alpha$**: Bác bỏ giả thuyết $H_0$. Kết luận hai lớp có sự khác biệt mang ý nghĩa thống kê về độ dài.
* Nếu **$p$-value $\ge$ $\alpha$**: Không có đủ bằng chứng để bác bỏ $H_0$. Kết luận hai lớp không có sự khác biệt mang ý nghĩa thống kê về độ dài (hay nói cách khác, độ dài không mang tính phân biệt nhãn).

```python
def plot_distribution(df, feature_col, target_col='sentiment'):
    """
    Hàm vẽ biểu đồ phân phối (Histogram) và Boxplot phân tách theo nhãn.

    Tham số:
    - df: DataFrame chứa dữ liệu.
    - feature_col: Cột dữ liệu muốn vẽ (ví dụ: 'word_count' hoặc 'character_length').
    - target_col: Cột chứa nhãn phân lớp (mặc định là 'sentiment').
    """
    # Khai báo màu sắc chuẩn để dùng chung cho cả 2 loại biểu đồ
    custom_palette = {'positive': 'coral', 'negative': 'dodgerblue'}

    # Định dạng lại tên cột để làm tiêu đề đẹp hơn (vd: 'word_count' -> 'Word Count')
    title_name = feature_col.replace('_', ' ').title()

    # ==========================================
    # FIGURE 1: HISTOGRAM CHO TỪNG LỚP
    # ==========================================
    # Khởi tạo khung hình với 1 hàng, 2 cột
    fig, axes = plt.subplots(1, 2, figsize=(18, 5))

    # 1. Histogram cho lớp Positive
    sns.histplot(
        data=df[df[target_col] == 'positive'],
        x=feature_col,
        color=custom_palette['positive'],
        kde=True,
        ax=axes[0]
    )
    axes[0].set_title(f'{title_name}: Positive', fontsize=18, fontweight='bold')

    # 2. Histogram cho lớp Negative
    sns.histplot(
        data=df[df[target_col] == 'negative'],
        x=feature_col,
        color=custom_palette['negative'],
        kde=True,
        ax=axes[1]
    )
    axes[1].set_title(f'{title_name}: Negative', fontsize=18, fontweight='bold')

    plt.tight_layout()
    plt.show()

    # ==========================================
    # FIGURE 2: BOXPLOT NẰM RIÊNG Ở DƯỚI
    # ==========================================
    plt.figure(figsize=(10, 4))

    sns.boxplot(
        data=df,
        x=target_col,
        y=feature_col,
        hue=target_col,
        palette=custom_palette,
        legend=False
    )
    plt.title(f'Boxplot of {title_name}', fontsize=16, fontweight='bold')

    plt.tight_layout()
    plt.show()
```

```python
def display_descriptive_statistics(df, feature_col, target_col='sentiment'):
    """
    Hàm hiển thị báo cáo thống kê định lượng dưới dạng Bảng HTML đẹp mắt.
    """
    title_name = feature_col.replace('_', ' ').upper()

    display(HTML(f"<h3 style='color: white;'>THỐNG KÊ MÔ TẢ: {title_name}</h3>"))

    # display(HTML("<b>1. Tổng quan</b>"))
    descriptive_statistics = df.groupby(target_col)[feature_col].describe().round(2)
    display(descriptive_statistics)

    display(HTML("<hr>"))
```

```python
def display_inferential_statistics(df, feature_col, target_col='sentiment'):
    """
    Hàm thực hiện và in báo cáo kiểm định Mann-Whitney U cho 2 lớp độc lập.
    """
    labels = df[target_col].unique()
    title_name = feature_col.replace('_', ' ').upper()
    display(HTML(f"<h3 style='color: white;'>THỐNG KÊ SUY DIỄN: {title_name}</h3>"))

    if len(labels) != 2:
        print("[WARN] Kiểm định này chỉ áp dụng cho 2 lớp phân biệt!")
        return

    label_1, label_2 = labels[0], labels[1]

    group_1 = df[df[target_col] == label_1][feature_col]
    group_2 = df[df[target_col] == label_2][feature_col]

    u_stat, p_val = stats.mannwhitneyu(group_1, group_2, alternative='two-sided')
    feature_name = feature_col.replace('_', ' ').upper()
    print(f" Kiểm định Mann-Whiteney U: {feature_name}")
    print(f"   • Giả thuyết H0  : Không có sự khác biệt về {feature_name.lower()} giữa '{label_1}' và '{label_2}'.")
    print(f"   • Kích thước mẫu lớp '{label_1}' : {len(group_1):,}")
    print(f"   • Kích thước mẫu lớp '{label_2}' : {len(group_2):,}")
    print(f"   (+) U-Statistic   : {u_stat:,.2f}")
    print(f"   (+) P-value       : {p_val:.4e}")

    alpha = 0.05
    if p_val < alpha:
        print(f"   => Kết luận: Bác bỏ H0 vì P-value < {alpha}.")
        print(f"      Có sự khác biệt mang ý nghĩa thống kê về {feature_name.lower()} giữa 2 lớp.")
    else:
        print(f"   => Kết luận: Không bác bỏ H0 do P-value >= {alpha}.")
        print(f"      Không có sự khác biệt mang ý nghĩa thống kê về {feature_name.lower()} giữa 2 lớp.")
    # print("_"*65 + "\n")
```

```python
def inspect_outliers(df, feature_col, text_col='review', target_col='sentiment', high_thresh=None, low_thresh=None, n_samples=3):
    """
    Hàm truy xuất và hiển thị nội dung của các văn bản ngoại lai (quá dài hoặc quá ngắn).
    Sử dụng print() thuần túy.
    """
    feature_name = feature_col.replace('_', ' ').upper()

    print(f"Sanity check outliers: {feature_name}")

    # 1. KIỂM TRA OUTLIERS QUÁ DÀI (High Threshold)
    if high_thresh:
        long_outliers = df[df[feature_col] > high_thresh].sort_values(by=feature_col, ascending=False)
        print(f"\nReview quá dài (>{high_thresh}) - Tổng số: {len(long_outliers)} mẫu")

        if not long_outliers.empty:
            # Lấy n_samples dài nhất
            for i, (_, row) in enumerate(long_outliers.head(n_samples).iterrows(), 1):
                label = str(row[target_col]).upper()
                length = row[feature_col]
                text = row[text_col]

                # In ra 500 ký tự đầu và 500 ký tự cuối để kiểm tra xem có kịch bản/HTML lặp lại không
                preview = f"{text[:500]} \n\n[ ... ] \n\n{text[-500:]}" if len(text) > 1000 else text

                print(f"Mẫu Dữ liệu {i} | Lớp Phân loại: {label} | Độ dài: {length:,}")
                print(preview)

    print("_" * 100)

    # 2. KIỂM TRA OUTLIERS QUÁ NGẮN (Low Threshold)
    if low_thresh:
        short_outliers = df[df[feature_col] < low_thresh].sort_values(by=feature_col, ascending=True)
        print(f"\nReview quá ngắn (<{low_thresh}) - Tổng số: {len(short_outliers)} mẫu")

        if not short_outliers.empty:
            # Lấy n_samples ngắn nhất
            for i, (_, row) in enumerate(short_outliers.head(n_samples).iterrows(), 1):
                label = str(row[target_col]).upper()
                length = row[feature_col]
                text = row[text_col]

                print(f"Mẫu Dữ liệu {i} | Lớp Phân loại: {label} | Độ dài: {length:,}")
                print(text)
                # print("_" * 80)

    print("\n")
```

### 4. Định hướng Tiền xử lý và Phân tích tiếp theo
Việc đặc trưng này vô giá trị trong việc phân loại không có nghĩa là nó vô giá trị trong quy trình. Các con số thống kê trên cung cấp các "hằng số" quan trọng cho khâu Data Cleaning:

2. **Cắt tỉa văn bản (Truncation / Padding Strategy):** Quan sát vào bách phân vị thứ 75 (Q3), 75% dữ liệu đều có số ký tự dưới mức **~1614**. Đây là một sweet point.

### Cơ sở Lý thuyết về Phân tích Tần suất Từ vựng (Word Frequency Analysis)

Trước khi mô hình học máy có thể hiểu được văn bản, chúng ta cần đánh giá xem những khái niệm nào (thông qua từ vựng) đang chiếm ưu thế trong không gian dữ liệu. Quá trình này dựa trên mô hình **Bag-of-Words (BoW)**, nơi trật tự từ bị bỏ qua, chỉ giữ lại sự xuất hiện và tần suất của chúng.

#### 1. Tần suất từ (Term Frequency) và Bảng Top-N
* **Cơ sở lý thuyết:** Tần suất từ đo lường mức độ phổ biến của một khái niệm. Trong bài toán phân tích quan điểm (Sentiment Analysis), việc so sánh các từ có tần suất cao nhất (Top-N) giữa lớp Positive và Negative giúp ta phác họa được "từ điển cảm xúc" đặc trưng của từng nhóm người dùng.
* **Định nghĩa toán học:** Giả sử tập ngữ liệu (Corpus) $C$ của một lớp (ví dụ: lớp Positive) bao gồm $M$ văn bản: $C = \{d_1, d_2, \dots, d_M\}$.
Gọi $f_{w, d}$ là số lần từ $w$ xuất hiện trong văn bản $d$.
Tần suất tổng cộng (Total Term Frequency) của từ $w$ trong toàn bộ ngữ liệu $C$ được tính bằng:
$$TF(w, C) = \sum_{d \in C} f_{w, d}$$
Bảng **Top 50 từ phổ biến nhất** chính là tập hợp 50 từ $w$ có giá trị $TF(w, C)$ cao nhất, được sắp xếp theo thứ tự giảm dần.

#### 2. Đám mây từ vựng (Word Cloud)
* **Cơ sở lý thuyết:** Bảng số liệu khô khan (Top 50) rất tốt cho máy tính, nhưng não người tiếp thu hình ảnh tốt hơn. Word Cloud là một phương pháp trực quan hóa dữ liệu không gian văn bản (Text Spatial Visualization).
* **Cơ chế hoạt động:** Trọng lượng thị giác (Visual Weight - cụ thể là cỡ chữ - Font Size) của mỗi từ $w$ trên biểu đồ tỷ lệ thuận với tần suất $TF(w, C)$ của nó. Từ nào xuất hiện càng nhiều, nó sẽ được vẽ ở vị trí trung tâm với kích thước càng lớn, giúp người phân tích nhận diện ngay lập tức "từ khóa" của tập dữ liệu.


### Cơ sở Lý thuyết về Chỉ số Phong phú Từ vựng (Type-Token Ratio)

Nếu Tần suất từ (TF) cho biết *review đang trình bày về điều gì*, thì Chỉ số phong phú từ vựng cho biết *trong review, người dùng nói điều đó với vốn từ đa dạng như thế nào*.

#### 1. Định nghĩa Token và Type
Để tính toán độ phong phú, ngôn ngữ học thống kê phân biệt rõ hai khái niệm:
* **Token (Mẫu vị):** Tổng số từ được viết ra. Ví dụ, câu *"to be or not to be"* có **6 tokens**.
* **Type (Kiểu vị/Từ vựng độc nhất):** Số lượng các từ khác nhau được sử dụng. Trong câu *"to be or not to be"*, chỉ có 4 loại từ khác nhau là {*to, be, or, not*}. Vậy có **4 types**.

#### 2. Công thức Type-Token Ratio (TTR)
* **Cơ sở lý thuyết:** TTR đo lường mức độ đa dạng từ vựng (Lexical Diversity) của một tập văn bản. Chỉ số này đặc biệt hữu ích để kiểm tra xem liệu người dùng đang bộc lộ cảm xúc bằng một tập hợp từ phong phú, hay chỉ dùng đi dùng lại một vài từ lóng/chửi thề rập khuôn.
* **Định nghĩa toán học:** Gọi $N$ là tổng số Tokens trong toàn bộ ngữ liệu $C$.
Gọi $V$ là tổng số Types (kích thước của từ điển từ vựng độc nhất - Vocabulary Size).
Chỉ số TTR được tính bằng tỷ lệ:
$$TTR = \frac{V}{N}$$
* **Tính chất:** TTR luôn nằm trong khoảng $(0, 1]$.
    * TTR tiến gần về $1$: Vốn từ cực kỳ phong phú (mỗi từ xuất hiện gần như chỉ một lần).
    * TTR tiến gần về $0$: Vốn từ rất nghèo nàn, lặp lại nhiều.

#### 3. Định luật Heaps
* Khi phân tích TTR, cần đối chiếu với **Định luật Heaps (Heaps' Law)**. Định luật này phát biểu rằng: khi kích thước văn bản ($N$) càng tăng, số lượng từ mới ($V$) sẽ tăng chậm dần (tuân theo hàm mũ $V = k \cdot N^\beta$ với $\beta < 1$).
* Do đó, với tập dữ liệu khổng lồ như IMDB (hàng triệu tokens), TTR thường sẽ hội tụ về một con số rất nhỏ (ví dụ $< 0.05$). Để so sánh công bằng, ta chỉ nên so sánh TTR giữa các lớp (Positive vs Negative) khi kích thước mẫu (tổng số tokens) của chúng xấp xỉ nhau.


### Tiền đề bắt buộc
Để các công thức toán học trên phản ánh đúng thực tế, toàn bộ quá trình đếm $TF$ và tính $TTR$ **bắt buộc phải được thực hiện sau bước Tiền xử lý cơ bản (Light Preprocessing)**:
1. **Chuẩn hóa chữ hoa/thường (Lowercasing):** Giúp "Movie" và "movie" được đếm là cùng một $Type$.
2. **Loại bỏ từ dừng (Stopwords Removal):** Nếu không loại bỏ các từ chức năng như "the", "a", "is", các từ này sẽ vĩnh viễn chiếm Top 1 đến Top 50 của mọi văn bản tiếng Anh, làm mất hoàn toàn giá trị của Word Cloud và sai lệch chỉ số TTR.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import nltk

# Tải các gói cần thiết (chỉ chạy 1 lần)
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Khởi tạo stopwords 1 lần (tránh gọi lại nhiều lần)
# Bổ sung các từ đặc thù của lĩnh vực điện ảnh (không mang ý nghĩa cảm xúc)
# và các từ xuất hiện quá nhiều nhưng không giúp ích cho phân tích
STOPWORDS = set(stopwords.words('english')).union(
    {'film', 'films', 'movie', 'movies', 'one', 'would', 'even', 'get', 'make', 'really', 'also', 'see', 'time', 'story'}
)

def get_clean_tokens(text, stop_words=STOPWORDS):
    """
    Hàm Light Cleaning dành riêng cho bước Khám phá dữ liệu (EDA).
    """
    text = str(text)
    text = re.sub(r'<.*?>', ' ', text.lower())          # Thay HTML bằng khoảng trắng (tránh dính chữ)
    text = re.sub(r'[^a-z\s]', ' ', text)               # Thay ký tự đặc biệt bằng khoảng trắng
    tokens = word_tokenize(text)

    # Lọc stopwords và những từ rác có 1 ký tự (như 's', 't', 'm' do lỗi tách chữ)
    return [t for t in tokens if t not in stop_words and len(t) > 1]


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
from IPython.display import display, HTML # THÊM THƯ VIỆN NÀY

def analyze_single_class_vocabulary(df, target_label, text_col='review', target_col='sentiment', top_n=50):
    """
    Hàm phân tích từ vựng (TTR, Top N từ, Word Cloud) cho MỘT LỚP CỤ THỂ,
    sử dụng display() để hiển thị giao diện đẹp mắt trên Jupyter Notebook.
    """
    # ========================================================
    # 1. TIÊU ĐỀ BÁO CÁO (DÙNG PRINT)
    # ========================================================
    print("\n" + "_"*75)
    print(f"PHÂN TÍCH TỪ VỰNG: LỚP [{str(target_label).upper()}]")
    print("_"*75)

    # Lọc dữ liệu
    subset = df[df[target_col] == target_label]
    if subset.empty:
        print(f"[WARN] Không tìm thấy dữ liệu nào thuộc lớp '{target_label}'!")
        return

    # Tiến hành Tokenize (Giả sử bạn đã định nghĩa hàm get_clean_tokens ở ngoài)
    all_tokens = [t for sublist in subset[text_col].apply(get_clean_tokens) for t in sublist]

    if not all_tokens:
        print("-> Không có từ nào sau khi làm sạch!")
        return

    # ========================================================
    # 2. CHỈ SỐ ĐỊNH LƯỢNG (TTR) - DÙNG PRINT
    # ========================================================
    total_tokens = len(all_tokens)
    unique_types = len(set(all_tokens))
    ttr = unique_types / total_tokens if total_tokens > 0 else 0

    print("\n1. Thống kê Định lượng độ phong phú Từ vựng")
    print(f"   • Tổng số từ vựng (Tokens)     : {total_tokens:,}")
    print(f"   • Tổng số từ vựng độc nhất (Types)  : {unique_types:,}")
    print(f"   • Chỉ số TTR (Type-Token Ratio): {ttr:.4f}\n")

    # ========================================================
    # 3. BẢNG TOP N TỪ PHỔ BIẾN NHẤT
    # ========================================================
    print(f"\n2. Bảng xếp hạng top {top_n} Từ vựng phổ biến nhất")

    counts = Counter(all_tokens)
    top_words = counts.most_common(top_n)

    # Đưa vào Pandas DataFrame
    df_top = pd.DataFrame(top_words, columns=['Từ vựng (Word)', 'Tần suất (Frequency)'])
    df_top.index = range(1, len(df_top) + 1)
    df_top.index.name = 'Rank'

    # Định dạng lại cột tần suất có dấu phẩy cho dễ đọc (VD: 39437 -> 39,437)
    df_top['Tần suất (Frequency)'] = df_top['Tần suất (Frequency)'].apply(lambda x: f"{x:,}")

    # CHỈ CẦN GỌI DISPLAY - PANDAS SẼ TỰ RENDER RA BẢNG HTML CÓ CHIA CỘT TUYỆT ĐẸP
    display(df_top)

    # ========================================================
    # 4. VẼ WORD CLOUD
    # ========================================================
    # display(HTML("<b>3. Đám mây từ vựng (Word Cloud):</b>"))

    print("\n3. Đám mây Từ vựng (Word Cloud)")

    plt.figure(figsize=(10, 5))
    cmap = 'viridis' if str(target_label).lower() == 'positive' else 'magma'

    wc = WordCloud(
        width=800,
        height=400,
        background_color='white',
        colormap=cmap,
        max_words=200
    ).generate_from_frequencies(counts)

    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')

    # Bạn thậm chí không cần plt.title() nữa vì đã có tiêu đề HTML ở trên,
    # nhưng giữ lại cho biểu đồ đầy đủ thông tin khi lưu ảnh
    plt.title(f'Word Cloud - {str(target_label).upper()}', fontsize=16, fontweight='bold', pad=15)

    plt.tight_layout()
    plt.show() # Matplotlib tự động tương thích rất tốt với IPython notebook

    # Đường phân cách kết thúc báo cáo
    display(HTML("<hr style='border-top: 2px solid #bbb;'>"))
```

```python
analyze_single_class_vocabulary(df, target_label='positive')
```

## Phân tích **phân phối Zipf**: vẽ log-log plot của tần suất từ và kiểm tra mức độ tuân theo định luật Zipf.


Phân phối Zipf (Zipf's Law) là một định luật thực nghiệm mang tính nền tảng, có giá trị cốt lõi trong phân tích định lượng học và đặc biệt quan trọng khi làm việc với các hệ thống Xử lý Ngôn ngữ Tự nhiên (NLP) và mô hình không gian vector (Vector Space Models). Định luật này mô tả mối quan hệ nghịch đảo mang tính quy luật giữa tần suất xuất hiện của một đối tượng và thứ hạng của nó trong một tập hợp.

### 1. Cơ sở lý thuyết
Được nhà ngôn ngữ học George Kingsley Zipf phổ biến vào những năm 1930, định luật này xuất phát từ "Nguyên lý nỗ lực tối thiểu" (Principle of Least Effort). Trong ngôn ngữ tự nhiên, hệ thống giao tiếp luôn tự cân bằng giữa hai thái cực:
* **Người nói** muốn dùng một số lượng ít các từ ngắn, đa nghĩa, lặp đi lặp lại để tối thiểu hóa nỗ lực phát âm và truy xuất trí nhớ.
* **Người nghe** lại cần một hệ thống từ vựng rộng lớn, chi tiết và đơn nghĩa để loại bỏ sự mơ hồ.

Sự thỏa hiệp mang tính tiến hóa này tạo ra một phân phối dạng đuôi dài (heavy-tailed distribution), nơi một số ít phần tử thống trị hoàn toàn về tần suất, trong khi đại đa số các phần tử khác xuất hiện rất hiếm hoi.

### 2. Công thức toán học
Giả sử chúng ta có một tập ngữ liệu (corpus) chứa một tập từ vựng (vocabulary) với kích thước $N$. Nếu ta đếm tần suất của từng từ và sắp xếp chúng theo thứ tự giảm dần, từ xuất hiện nhiều nhất sẽ có thứ hạng $r = 1$, từ tiếp theo có $r = 2$, cho đến $r = N$.

Định luật Zipf phát biểu rằng tần suất $f(r)$ của từ có thứ hạng $r$ tỷ lệ nghịch với $r$ lũy thừa $s$:
$f(r) \propto \frac{1}{r^s}$

Trong đó $s$ là một tham số thực đặc trưng cho tập dữ liệu (với ngôn ngữ tự nhiên, $s$ thường xấp xỉ bằng $1$).

Xác suất xuất hiện $P(r)$ của từ ở thứ hạng $r$ được tính bằng cách chuẩn hóa trên toàn bộ không gian mẫu:
$P(r) = \frac{\frac{1}{r^s}}{\sum_{n=1}^N (\frac{1}{n^s})}$

Đại lượng ở mẫu số $\sum_{n=1}^N (\frac{1}{n^s})$ chính là số Harmonic tổng quát (Generalized Harmonic Number).

**Hệ thức Logarit:** Để kiểm chứng phân phối Zipf, các nhà khoa học dữ liệu thường chuyển đổi phương trình sang hệ trục tọa độ logarit (log-log scale). Lấy logarit cơ số tự nhiên hai vế, ta được:
$\log(f(r)) = C - s \cdot \log(r)$
(Với $C$ là hằng số). Trên đồ thị log-log, phân phối Zipf hoàn hảo sẽ hiển thị dưới dạng một đường thẳng tuyến tính dốc xuống với hệ số góc là $-s$.

### 3. Ví dụ thực tiễn
Trong một ngữ liệu tiếng Anh quy mô lớn, phân phối Zipf thể hiện rất rõ ràng và ổn định:
* Từ hạng 1 là **"the"**, thường chiếm khoảng 7% tổng số lượng từ của toàn bộ văn bản.
* Từ hạng 2 là **"of"**, chiếm khoảng 3.5% (tức là $\frac{1}{2}$ tần suất của "the").
* Từ hạng 3 là **"and"**, chiếm khoảng 2.3% (tức là $\frac{1}{3}$ tần suất của "the").

**Ý nghĩa ứng dụng:**
Chính vì quy luật này, phần lớn dữ liệu văn bản mà các mô hình học máy tiếp xúc đều thuộc về the long tail (các từ rất hiếm). Đây là cơ sở khoa học cốt lõi cho các kỹ thuật tiền xử lý:
1.  **Loại bỏ Stop-words:** Các từ ở đầu phân phối (như "the", "an", "a", "is") xuất hiện quá nhiều nhưng mang lượng thông tin (Entropy) rất thấp, cần bị loại bỏ để giảm nhiễu.
2.  **Thuật toán TF-IDF:** Trọng số TF-IDF được sinh ra để phạt (penalize) các từ tuân thủ quá mạnh phân phối Zipf (chỉ số DF cao), từ đó làm nổi bật lên các từ vựng mang đặc trưng ngữ nghĩa riêng biệt của văn bản.

* **Hệ số R-squared (0.9845):** Mức độ khớp tuyến tính trên thang logarit lên tới 98.45% là một điểm số xuất sắc (vượt xa ngưỡng 0.95). Ta hoàn toàn tự tin kết luận dữ liệu tuân thủ nghiêm ngặt quy luật ngôn ngữ thống kê.
* **Hệ số Alpha (1.1124):** Trong một ngữ liệu tiếng Anh lý tưởng, $\alpha$ thường xấp xỉ 1.0. Việc $\alpha = 1.11$ ( với độ dốc gắt hơn một chút) cho thấy sự chênh lệch tần suất giữa các từ phổ biến và từ hiếm trong tập IMDB này diễn ra **nhanh hơn bình thường**. Điều này phản ánh tính chất "miền đặc thù" (Domain-specific): người dùng tập trung lặp lại một nhóm từ hẹp về điện ảnh, khiến các từ khác tụt hạng rất nhanh.

### 3. Vấn đề "Hapax Legomena" và Khoảng trống Tiền xử lý
Hãy tập trung chú ý vào hai con số:
* Tổng số từ vựng (Vocabulary Size): **214,473 từ**
* Số từ xuất hiện đúng 1 lần (Hapax Legomena): **133,030 từ (Chiếm tới 62%!)**

**Insight:** Việc có tới 62% lượng từ vựng chỉ xuất hiện đúng 1 lần trong suốt 50,000 bài review là một **Red Flag (Cờ đỏ)** cảnh báo lớn.
Tại sao lại có nhiều từ độc nhất đến vậy?
1.  **Lỗi chính tả (Typos):** Người dùng gõ sai (ví dụ: *awesommme*, *terribleee*).
2.  **Tên riêng:** Tên diễn viên vô danh, tên đạo diễn nước ngoài.
3.  **Lỗi Tokenize:** Các từ bị dính liền do thiếu dấu cách (ví dụ: *goodmovie*, *lovedit*).

### 4. Kết luận và đề xuất
Định luật Zipf không chỉ dùng để đánh giá chất lượng ngôn ngữ, mà các tham số hình học của nó sẽ trực tiếp định hướng cách chúng ta thiết kế không gian vector (Vectorization) ở bước tiền xử lý ngay sau đây:

#### Pruning the Tail
* Khối lượng 154,641 từ hiếm (xuất hiện < 3 lần) ở phần đuôi đồ thị thực chất là nhiễu (noise) như lỗi chính tả, tên riêng vô nghĩa. Nếu ném toàn bộ ma trận hơn 200,000 chiều này vào mô hình, ta sẽ gặp phải Lời nguyền Chiều dữ liệu (Curse of Dimensionality) gây tràn RAM.

* Đề xuất: Khi thiết lập bộ khởi tạo ma trận từ vựng (như TfidfVectorizer hoặc CountVectorizer), cần bắt buộc sử dụng tham số min_df (Minimum Document Frequency) với ngưỡng khoảng từ 5 đến 10. Thao tác này sẽ thẳng tay "chặt bỏ" cái đuôi khổng lồ, thu gọn không gian đặc trưng xuống chỉ còn khoảng 20,000 - 30,000 từ chất lượng nhất mà không làm hao hụt ngữ nghĩa.

#### Pruning the Head:
* Các từ ở chóp đỉnh (như movie, film) có tần suất khổng lồ và xuất hiện ở hầu hết mọi văn bản, đóng vai trò như các "Stopwords miền đặc thù" (Domain Stopwords).

* Đề xuất: Thay vì phải cất công dò tìm và tự lập danh sách Stopwords điện ảnh thủ công, ta có thể tự động hóa việc "cắt đỉnh" bằng cách thiết lập tham số max_df = 0.85 (loại bỏ các từ xuất hiện ở hơn 85% tổng số văn bản). Điều này nhằm giúp triệt tiêu nhiễu và làm nổi bật các từ và cụm từ mang sắc thái cảm xúc.


## Tổng kết (EDA) & Đề xuất cho Tiền xử lý

### Tổng kết các phát hiện cốt lõi (Key Findings)

Trong quá trình phân tích phân phối độ dài, từ vựng và định luật Zipf, chúng ta đã nắm bắt được 4 vấn đề cốt lõi của tập dữ liệu:

1.  **Nhiễu Kỹ thuật (Technical Noise):** Việc kiểm tra Outliers tiết lộ phần đuôi của dữ liệu chứa các bài đánh giá siêu dài bị lấp đầy bởi các thẻ HTML rác (đặc biệt là `<br /><br />`), các URL, và dấu câu vô nghĩa.
2.  **Sự phân mảnh Từ vựng (Word Fragmentation):** Thông qua Word Cloud và bảng Top 50, ta thấy các từ mang cùng ngữ nghĩa đang bị máy tính tách thành các đặc trưng riêng biệt (VD: `movie` - `movies`, `film` - `films`, `character` - `characters`).
3.  **(Domain Stopwords):** Mặc dù đã loại bỏ stopwords cơ bản của NLTK, các từ như `movie`, `film`, `one`, `story` vẫn chiếm sóng áp đảo. Chúng không mang lại giá trị phân loại cảm xúc (Information Gain) mà chỉ làm loãng ma trận.
4.  **Lời nguyền Chiều dữ liệu (Curse of Dimensionality):** Phân tích Zipf's Law chỉ ra một sự thật giật mình: Có tới hơn 214,000 từ vựng độc nhất, nhưng 62% trong số đó (Hapax Legomena) chỉ xuất hiện đúng 1 lần (do lỗi chính tả, từ ghép bừa bãi). Nếu giữ nguyên, mô hình sẽ bị Overfitting và tràn RAM.


### Đề xuất cho Tiền xử lý (Proposed Pipeline)

Để giải quyết triệt để 4 vấn đề trên, ta đề xuất thiết kế quy trình thực nghiệm bám sát đúng 6 yêu cầu Data preprocessing trong đồ án:

#### Pipeline chuẩn hóa Văn bản
* **Hành động:** Giải quyết "Nhiễu Kỹ thuật". Xây dựng một hàm Regex mạnh mẽ để triệt tiêu `<br>`, URL, và chuyển `lowercase`.
* **Đề xuất:** Thực thi pipeline này và đo lường sự sụt giảm của "Số từ/Số ký tự tối đa" (giảm các Outliers 13,000 ký tự xuống mức thực tế hơn).

#### So sánh Chiến lược Tokenization
* **Hành động:** Giải quyết bài toán cắt từ.
* **Đề xuất thực nghiệm:** Khi so sánh Word-level (NLTK) với Subword (BPE), BPE sẽ chứng minh được sức mạnh trong việc giải quyết 62% từ hiếm (Hapax Legomena) ở phần đuôi Zipf bằng cách tách các từ sai chính tả hoặc từ ghép thành các subword phổ biến hơn, qua đó giảm mạnh tỉ lệ OOV (Out-of-Vocabulary).

#### 3. Tối ưu hóa Không gian Từ vựng
* **Loại bỏ Stopwords:** Dùng Mutual Information (MI) để chứng minh bằng toán học rằng các từ `movie` hay `film` có điểm MI cực thấp với nhãn (không có tính phân biệt), từ đó biện luận việc xóa bỏ chúng.
* **Stemming vs. Lemmatization:** Giải quyết "Sự phân mảnh từ vựng". Tính Collision Rate để xem phương pháp nào gộp `movies` về `movie` hiệu quả nhất. *Nhận định:* WordNet Lemmatizer thường chạy chậm hơn nhưng an toàn cho ngữ nghĩa hơn Porter Stemmer.

#### Vector hóa và Ngữ nghĩa
* **Xử lý Lời nguyền Chiều dữ liệu:** Khi cấu hình TF-IDF, ta bắt buộc phải dùng tham số `min_df` (để cắt đuôi Zipf) và `max_df` (để cắt chóp Zipf). Việc dùng N-gram (bi-gram, tri-gram) là bắt buộc để mô hình học được hiện tượng "Nghịch lý từ phủ định" (VD: `not good`, `never watched`) đã phát hiện ở EDA.
* **Word2Vec & Sentence Transformer:** So sánh sức mạnh của ma trận thưa (TF-IDF với Sparsity Ratio > 99%) với ma trận dày đặc (Embeddings) khi đã thực hiện preprocessing.
